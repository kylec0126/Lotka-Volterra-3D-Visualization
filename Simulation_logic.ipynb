```python
import numpy as np
from scipy.integrate import odeint
import pandas as pd

# 1. Define the Lotka-Volterra equations
def deriv(state, t, alpha, beta, delta, gamma):
    x, y = state
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]

# 2. Parameters
alpha, beta = 1.1, 0.4
delta, gamma = 0.1, 0.4
t = np.linspace(0, 50, 1000) # 1000 time steps

# 3. Generate several trajectories (Parameter Sweep)
# We will vary the initial population of Prey
all_data = []
for initial_prey in range(5, 25, 5):
    initial_state = [initial_prey, 5]
    sol = odeint(deriv, initial_state, t, args=(alpha, beta, delta, gamma))
    
    df = pd.DataFrame(sol, columns=['Prey', 'Predator'])
    df['Time'] = t
    df['Initial_Prey'] = initial_prey # Our 3rd dimension/parameter
    all_data.append(df)

# 4. Combine and Export
final_df = pd.concat(all_data)
final_df.to_csv("lotka_volterra.csv", index=False)
print("Data exported to lotka_volterra.csv")
```
